### Transform
We used Pandas and Python in order to clean up our data. For our csv, we did the following:
1. Use groupby function in order to get the types of pokemons.
2. Added a None category for Pokemons without a secondary type.
3. Rename our column names so that the would match the tables in our database.
4. Rename the index to type_id.

For our JSON files, we did the following:
1. Drop columns that we didn't need.
2. Drop all types of forms, except Normal.
3. Create a function to split the column types into types and sub-types.
4. Create a function that checks if new types have been added.
5. Merge tables.
6. Drop multiple copies of Arceus from final table.
7. Check if there is new data by comparing 2 Dataframes.

```python
# Groupby main type
df_2 = df_2.groupby('Type 1').mean()
df_2
```

```python
df_2 =  df_2.reset_index()
df_2
```

```python
# Add 'None' type for pokemon without a sub-type
df_2 = df_2[['Type 1']]
df_2
```

```python
# Rename column so they match pgAdmin table
df_2 = df_2.rename(columns={"Type 1": "types"})
df_2.head()
```

```python
# Rename index
df_2.index.name = 'type_id'
df_2 = df_2.reset_index()
```

```python
df_2.head()
```

```python
# Drop duplicates and unnecesarry columns
df_3 = df_3.drop_duplicates(subset="pokemon_name")
df_3 = df_3[["base_attack", "base_defense", "base_stamina", "pokemon_id"]]
df_3
```

```python
# Drop all forms except 'Normal'
types = df_2['types'].values
df['check'] = df['form'].isin(types)
df.head()
```

```python
df = df.loc[df['check'] == True]
df.head()
```

```python
df = df.drop(columns=['check'])
df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
# Create function to split 'type' column into 'type' and 'sub-type'
def sub_type_formula(x):
    if len(x) > 1:
        return x[1]
    else:
        return 'None'
    
def type_formula(x):
    return x[0]

df['sub-type'] = df.apply(lambda row: sub_type_formula(row['type']), axis=1)
df['type'] = df.apply(lambda row: type_formula(row['type']), axis=1)
df
```

```python
# Check for new types of Pokemon
current_types = df_2['types'].tolist()
all_types = df['type'].tolist()
all_sub_types = df['sub-type'].tolist()

for poke_type in all_types:
    if poke_type not in all_types:
        pikachu = {"type_id": len(current_types), "types": [poke_type]}
        temp_df = pd.DataFrame(data=pikachu)
        df_2 = df_2.append(temp_df, ignore_index=True)
        all_types.append(poke_type)

for poke_type in all_sub_types:
    if poke_type not in all_types:
        pikachu = {"type_id": len(current_types), "types": [poke_type]}
        temp_df = pd.DataFrame(data=pikachu)
        df_2 = df_2.append(temp_df, ignore_index=True)
        all_types.append(poke_type)
        
df_2
```

```python
# Merge to add pokemon_type_id
df = df.merge(df_2, left_on="type", right_on="types")
df = df.rename(columns={"type_id": "pokemon_type_id"})
df
```

```python
# Adding subtype id
df = df.merge(df_2, how="left", left_on="sub-type", right_on="types")
df = df.rename(columns={"type_id": "pokemon_sub-type_id"})
df
```

```python
# Merge to add stats
df = df.merge(df_3, how="left", left_on="pokemon_id", right_on="pokemon_id")
df
```

```python
# Include columns needed for pgAdmin
column_names = ["pokemon_id", "pokemon_type_id", "pokemon_sub-type_id", "pokemon_name", 
                "base_attack", "base_defense", "base_stamina"]
df = df[column_names]
df
```

```python
# Drop many instances of Arceus
df.drop(df.loc[(df.pokemon_name == 'Arceus') & (df.pokemon_type_id != 12 )].index, inplace=True)
df
```

### Load

#### Create database connection

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/Pokemon_DB')
connection = engine.connect()
```

```python
# Confirm tables
engine.table_names()
```

#### Finding if there is new data to append

```python
current_types = pd.read_sql_query('select * from "types"',con=engine)
current_types
```

```python
new_types = pd.concat([current_types, df_2]).loc[current_types.index.symmetric_difference(df_2.index)]
new_types
```

```python
current_pokemon = pd.read_sql_query('select * from "pokemon"',con=engine)
current_pokemon
```

```python
df = df.reset_index()
new_pokemon = pd.concat([current_pokemon, df]).loc[current_pokemon.index.symmetric_difference(df.index)]
new_pokemon = new_pokemon.drop(columns=['index'])
new_pokemon
```

#### Load DataFrames into database

```python
new_types.to_sql(name='types', con=engine, if_exists='append', index=False)
```

```python
new_pokemon.to_sql(name='pokemon', con=engine, if_exists='append', index=False)
```
